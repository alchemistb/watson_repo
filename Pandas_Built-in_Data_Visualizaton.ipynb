<h3>To call a line plot graph with a df</h3> 

```python
# Usage: some df.plot.line(x=some index,y='some column')

df1.plot.line(x=df1.index,y='B')
```

```python
# Customize the line plot
# Change the figsize
# Change the linewidth

df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

<h3>To call a scatter plot graph of a df</h3>

```python
# Usage: some df.plot.scatter(x='some column1', y='some column2')

df1.plot.scatter(x='A',y='B')
```

```python
# Change the color for another column variable
# Add another column ( c='some column3')
# Use cmap='some cmap color' (i.e.,seismic, coolwarm, magma, etc.)

df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
# Instead of adding a color, use a size (s)
# Usage: s=some df['some column']*100


df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```
