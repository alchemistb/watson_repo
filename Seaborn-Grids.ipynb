<h1>Seaborn-Grids</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load Dataset iris
iris = sns.load_dataset('iris')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris['species'].unique()    # Show the three unique species
```

<h3>PairGrids</h3>

```python
# First, call pairplot() as a refresher


sns.pairplot(iris)
```

```python
# PairGrid()

sns.PairGrid(iris)   # Creates empty grids, but similar display to pairplot()
```

```python
# Can use the PairGrid as a 'blank canvas',
# and map plot types.

# Example, first call PairGrid(df), and assign to a variable
# then map the variable to a scatter plot

g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Various map plot types examples:

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)    # histogram plot on diagonal grids
g.map_upper(plt.scatter)    # scatter plot on upper grids
g.map_lower(sns.kdeplot)    # kde (density pattern) on lower grids
```

<h3>Facet Grid</h3>

```python
# Load tips dataset

tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
# Usage:
# sns.FacetGrid(data=some data, col='some column to use as header', row= 'some column to use as row')

g = sns.FacetGrid(data=tips,col='time',row='smoker')
```

```python
# Facet Grid

# .FacetGrid(data=some data, col='some column to use as header', row= 'some column to use as row')
# .map(some type plot, 'some other column')

# Map a plot type(from .map) and separating results by column and row (from FacetGrid),
# Row = smoker (yes or no)
# Column = time (Lunch or Dinner)
# X-axis = total_bill (from .map)

g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')     # historgram plot type
```

```python
# If the .map needs more than one parameter,
# simply add it


# Example in .map, when using a plt. scatter, two arguments are needed,
# add 'tip' along with 'total_bill'


g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')    # Scatter plot uses x,y data
```
