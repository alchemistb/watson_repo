<h1>Seaborn-Matrix Plots</h1>

```python
import seaborn as sns
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

<h2>Heat Maps</h2>

```python
tc = tips.corr()    # The columns and rows need to be in a matrix format,
                    # using correlation corr() in order to use a heat map
```

```python
tc
```

```python
sns.heatmap(tc)     # Heatmap Usage:
                    # sns.heatmap(some correlated df/matrix)
```

<h3> The heat map shows different colors of values based on a gradient scale </h3>

```python
tc  # The values in the tc matrix correlate to the colors in the above heatmap
```

```python
sns.heatmap(tc,annot=True)  # Annotate to display matrix values in each heatmap square
                            # Usage: annot=True
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')  # Provide color to the heatmap
                                            # Usage: cmap='some color'
```

```python
flights
```

```python
# Create a matrix table for flights using pivot_table(), to be used in a heatmap

# Usage: df.pivot_table(index='some column to use as index', 
# columns='some column as column headers', 
# values='some column to populate the matrix with values')

fp = flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
fp.head()
```

```python
# Create a heatmap for flight (fp) matrix data

# Notice: Month (index) is on Y-axis, and year (column) is on X-axis
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='magma')    # cmap = 'magma' 
                                # The color goes from light to dark,
                                # for better readability
```

```python
# linecolor=white
# linewidths=1  (line thickness)
# To make the heatmap easier to read, and look sharp

sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

```python
# Another heatmap customization example:
# coolwarm cmap with black linecolor, and lw of 3

sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths=3)
```

<h2>Cluster Maps</h2>

<h3>Uses hierarchical clustering to produce a clustered version of a heatmap.</h3>
<h4>Grouping like things together</h4>

```python
sns.clustermap(fp)
```

```python
# Customizing a clustermap

sns.clustermap(fp,cmap='coolwarm',linecolor='black',lw=3)
```

```python
# Normalizing the scale to be between 0 - 1

sns.clustermap(fp,cmap='coolwarm',linecolor='black',lw=3,standard_scale=1)
```
