```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y = x ** 2
```

<h3>Functional method for creating plots</h3>


```python
plt.plot(x,y)
plt.xlabel('X Lablel')
plt.ylabel('Y Label')
plt.title('Title')
```

<h4>Adding multi plots</h4>

```python
# Subplot arguments take number of row, columns, and plot number reference

# Subplot #1
# X,Y, with a red line
plt.subplot(1,2,1)
plt.plot(x,y,'r')

# Subplot #2
# Y,X, with a blue line
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

<h3>Object Oriented method for creating plots</h3>


```python
fig = plt.figure()     # Create a figure object 'blank canvas'
```

```python
fig = plt.figure()     # Create a figure object


axes = fig.add_axes([0.1,0.1,0.8,0.8])  # Add a set of axes[LEFT, BOTTOM , WIDTH ,HEIGHT] to the figure (fig) object,
                                        # Using .add_axes()
    
    
axes.plot(x,y)                   # Plot the graph calling .plot() to axes
axes.set_xlabel('X Label')       # Set the X axis label, using set_xlabel()
axes.set_ylabel('Y Label')       # Set the Y axis label, using set_ylabel()
axes.set_title('Set Title')      # Set the title of the graph, using set_title()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  # The main box

#axes2 = fig.add_axes([0.2,0.2,0.4,0.3]) # Intial size of subplot box

axes2 = fig.add_axes([0.4,0.2,0.4,0.3])  # The inserted subplot box in bottom right corner
axes3 = fig.add_axes([0.5,0.3,0.2,0.1])  # The smallest box inside the inserted subplot box

axes1.plot(x,y)                   # plot lines on axes1
axes1.set_title('Larger Plot')    # Set title on axes1

axes2.plot(y,x)                   # plot lines on axes2
axes2.set_title('Smaller Plot')   # Set title on axes2

axes3.plot(x,x)
axes3.set_title('Tiny plot')      # plot lines on axes3
                                  # Set title on axes3 
```

<h1> Introduction to Matplotib - Part 2</h1>

<h2>Subplots</h2>

```python
fig,axes = plt.subplots(nrows=1, ncols=2)       # Create subplot 'blank canvas'
                                                # Assign fig and axes to plt.*subplots()* "Tuple Unpacking"
                                                # Use the .subplots()
                                                # nrows (number of rows)
                                                # ncols (number of columns)
     


                                                # **** To Plot x,y on the axes, 2 ways to do it: ****

                                                # ** ITERATION**
#for current_ax in axes:                         # Since axes is an array, it can be used in an iteration
    #current_ax.plot(x,y)                        # Using iteration (FOR Loop) to plot x and y along the axes,
                                                # in both subplots
                                                
                                                # -OR-
            
                                                # ** INDEXING**
axes[0].plot(x,y)                               # Using indexing to plot x,y in the first subplot (index 0)
axes[1].plot(y,x)                               # Using indexing to plot y,x in the second subplot (index 1)



                                                # ** SET LABELS**
axes[0].set_xlabel('X-label')                   # Set X axis label for axes[0]
axes[0].set_ylabel('Y-label')                   # Set Y axis label for axes[0]
axes[0].set_title('Subplot axes[0]')            # Set Title label for axes[0]

axes[1].set_xlabel('X-label')                   # Set X axis label for axes[1]
axes[1].set_ylabel('Y-label')                   # Set Y axis label for axes[1]
axes[1].set_title('Subplot axes[1]')            # Set Title label for axes[1]


plt.tight_layout()                             # To clean up the spaces so the subplots look less cluttered. 
```

<h2>Figure Size and DPI</h2>


```python
fig = plt.figure(figsize=(8,2))     # plt.figure(figsize=())
                                    # Pass in figsize('width', 'height') as a tuple,
                                    # *optional* dpi - dots per inch /pixels (usually just take the default)
    
    
ax = fig.add_axes([0,0,1,1])        # Add axes points ([left, bottom, width, height])

ax.plot(x,y)                        # Plot x,y points
```

<h3>Fig size and DPI with subplots</h3>

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))    # Fig and axes with plt.*suplots()*
                                                            # nrows, ncols,
                                                            # figsize('width','height')


axes[0].plot(x,y,label = 'Subplot1')                        # Using indexing to get the first subplot for x,y, and label for legend
axes[1].plot(y,x,label = 'Subplot2')                        # Using indexing to get the second subplot for y,x, and label for legend

axes[0].legend()                                            # Calling legend() to first subplot
axes[1].legend()                                            # Calling legend() to second subplot

plt.tight_layout()                                          # Clean up spacing between the subplots
```

<h2>Control over axis apperance</h2>

<h3>Plot range</h3>

<h3>We can configure the ranges of the axes using the <u>set_ylim</u> and <u>set_xlim</u> methods in the axis object, or <u>axis('tight')</u> for automatically getting "tightly fitted" axes ranges:</h3>

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')




ax.set_xlim([0,1])      # show the plot between 0 and 1 on the X axis
                        # .set_xlim(['some lower bound, some upper bound'])
    
ax.set_ylim([0,2])      # show the plot between 0 and 1 on the Y axis
                        # .set_ylim(['some lower bound, some upper bound'])
```
