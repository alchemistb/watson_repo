<h1>Seaborn-Categorical Plots</h1>

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

<h3>Bar Plot</h3>

```python
sns.barplot(x='sex',y='total_bill',data=tips)   # Barplot()
                                                # Used to compare data points from categorical and numerical columns
                                                # Usage: sns.barplot(x='[categorical column ]', y='[numerical column]', data=some df)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)  # Provide a way to get the estimate of the data ouput
                                                                # estimator = np.std (the standard deviation), applied to the barplot
```

<h3>Count Plot</h3>

```python
sns.countplot(x='sex',data=tips) # Similar to barplot, however,
                                 # The estimator is only counting the occurances.
                                 # Count is the 'Y' axis, which is automatically populated
                                 # similar to pandas.counts()
                                 # Usage: sns.countplot(x='some categorical column', data=some df)
```

<h3>Box Plots </h3>

<h4>Used to show the distribution of *categorical* data</h4>

```python
sns.boxplot(x='day',y='total_bill',data=tips)   # Usage: sns.boxplot(x='categorical column', y='numerical column', data=df)
                                                # The output is in qaurtiles 
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')  # Use hue to compare a specific column against,
                                                            # the x and y variables
                                                            # In this example, smoker is included (via hue) in the output
```

<h3>Violin Plots</h3>

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)    # This shows the KDE data distribution,
                                                    # where as boxplot does not
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')  # Include hue to get a more specific comparison
                                                            # In this example, we've determined the male/female
                                                            # breakdown of days of week vs. total_bill distribution
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True) 

# This provides more information by 'combing' 'sex',
# distributed across the day and total_bill data points for a more
# detailed informative graph
```

<h3>Strip Plots</h3>

```python
# Strip plot creates a scatterplot based off a category (x ='day')

sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
# jitter=True
# Jitter helps with the density of data by adding a little noise; 
# Data can be better seen and understood

sns.stripplot(x='day',y='total_bill', data=tips,jitter=True)
```

```python
# include hue='sex' (categorical column)
# Compare this column against the other variables

sns.stripplot(x='day',y='total_bill', data=tips,jitter=True,hue='sex')
```

```python
# Split to combine and compare the sex category with the disrtibuted variables

# Note: split is being replaced with 'dodge' in stripplot()

#sns.stripplot(x='day',y='total_bill', data=tips,jitter=True,hue='sex',split=True)
sns.stripplot(x='day',y='total_bill', data=tips,jitter=True,hue='sex',dodge=True)
```

<h3>Swarm Plots</h3>

```python
# Swarm plot is similar to stripplot(), 
# however, there is no overlapping
# This combines a stripplot with a violinplot

sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# It is possible to execute a violinplot and swarmplot as seen below
# However, it's not recommended when using large numbers/data sets,
# and it is very computationally intensive

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

<h3>Factor/Cat Plots</h3>

```python
# The most general form of distribution plots

# Usage: sns.factorplot(x='some categorical column', y='some numberical column', data=df, kind='what kind of plot')

# Note: factorplot is being replaced by catplot in future python releases

#sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```
