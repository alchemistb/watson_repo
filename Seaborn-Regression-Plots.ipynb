<h1>Seaborn: Regression-Plots</h1>
<h4>(LM) Linear Models</h4>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

<h3>LM Plots</h3>

```python
# To create a scatter plot with a linear fit

# Usage:
# sns.lmplot(x='some column', y='some column', data=df/dataset)

sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# Add hue to get seperation based off a categorical feature  
# This will produce two scatter plots and two linear points

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# Change marker types for the scatter plots
# markers=['some maker1','some marker2']
# Note: the number of markers must match the number of items assigned to hue
# hue = male,female - therefore 2 markers are needed ('o','v')


sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'])
```

```python
# To change the size of the markers, use scatter_kws={}
# scatter_kws={'s':some size}
# Note: s = size of a scatter plot

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'],
          scatter_kws={'s':100})
```

```python
# Can use a col instead of hue to separate categorical features
# This will create two columns/grids

# Example: Male and Female

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
# Can use row to add categorical features instead of hue
# This will create four grids

# Example: Lunch and Dinner

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
# Can include hue along with col and row
# Example:
# col=day (THUR,FRI,SAT,SUN)
# row=time (Lunch,Dinner)
# hue (Male,Female)


sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')
```

```python
# Drop row to only show day with male/female linear points

sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
# Can change the size of the grids using aspect and size
# aspect - height and width of each grid
# size - size of each grid

sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,size=8)
```
