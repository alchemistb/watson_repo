<h1>Seaborn - Style and Color</h1>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

<h2>Setting Styles</h2>

```python
# Creating set styles as follows:
# Usage: sns.set_style(style='some color')

#sns.set_style(style='white')
#sns.set_style(style='ticks')
#sns.set_style(style='darkgrid')
sns.set_style(style='whitegrid')


# Implementing a countplot graph is used to illustrate the set styles above
# Usage: sns.countplot(x='some column to use for an x-axis', data=df/dataset)

sns.countplot(x='sex',data=tips)
```

```python
# To remove the spines on a graph, not recommended

sns.despine()
sns.countplot(x='sex',data=tips)
```

<h2>Size and Aspect</h2>

```python
# To adjust the size and aspect of a graph,
# use plt.figure(figsize=(w,h))
# in combination with sns.countplot()

plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

<h2>Scale and Context</h2>

```python
# To change the graph context type to a poster, paper, etc.,
# use sns.set_context('context type')
# Use this with sns.countplot()

#sns.set_context('poster') # This sets the contextx to accomodate a poster
#sns.set_context('poster',font_scale=3) # This changes the font scale size on the graph

#sns.set_context('paper')   # This sets the context to accomodate a paper

sns.set_context('notebook') # This sets the context to accomodate a jupyter notebook

sns.countplot(x='sex',data=tips)
```

<h2>Palettes and Colors</h2>

```python
# Recall, using the lm plot to show colors, 
# with a hue for male/female

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# Use palette to change the color of the graph
# Example palette='seismic'

# References for palette colormaps: 
# https://matplotlib.org/examples/color/colormaps_reference.html

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
