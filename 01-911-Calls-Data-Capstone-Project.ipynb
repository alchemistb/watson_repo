** Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
# Use Conditional to find Traffic

df[df['Reason'] == 'Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
# Use Conditional to find Fire

df[df['Reason'] == 'Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
# Use Conditional to find EMS

df[df['Reason'] == 'EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
plt.tight_layout()
```

** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. There are lots of ways to do this, but I would recommend trying to combine groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method. Reference the solutions if you get stuck on this!**

```python
df.head()
```

```python
# 1 - Groupby multiple specific columns, using a list for the columns
# (i.e., df.groupby(by=['Day of Week','Hour']).count())

# 2 - Create the multilevel index by providing a column to match against
# (i.e., .count()['Reason'])

# 3 - Use the 'unstack' method
# (i.e., df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack())


# NOTICE: COULD ALSO USE PIVOT TABLE



df.groupby(by=['Day of Week','Hour']).count().head()
```

```python
df.groupby(by=['Day of Week','Hour']).count()['Reason'].head()
```

```python
#df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()

heat = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
```

```python
heat
```

** Now create a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.heatmap(heat,cmap='coolwarm')
plt.tight_layout()
```

** Now create a clustermap using this DataFrame. **

```python
sns.clustermap(heat,cmap='coolwarm')
plt.tight_layout()
```
